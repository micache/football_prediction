```python
import soccerdata as sd
import asyncio
import json
import aiohttp

from datetime import date
from understat import Understat

class InternalDataloader:

    fbref = None
    schedule = None
    league = None
    season = None

    def __init__(self, year, league):
        #self.fbref = sd.FBref(leagues=league, seasons=year)
        #self.schedule = self.fbref.read_schedule()
        self.season = year
        self.league = league

    def find_match_id(self, df, date, home):
        match = df.loc[(df['date'] == date) & ((df['home_team'] == home) | (df['home_team'] == home.replace('United', 'Utd'))), 'game_id']
        return match.iloc[0] if not match.empty else None

    def categorize_position(self, position):
        if not position or not isinstance(position, str):
            return "Unknown"
        
        last_char = position[-1]
        
        if last_char == "K":
            return "Goalkeeper"
        elif last_char == "B":
            return "Defender"
        elif last_char == "M":
            return "Midfielder"
        elif last_char == "W" or last_char == "F":
            return "Forward"
        
        return "Unknown"

    # Function to extract team structure
    def get_team_structure(self, df, team_name):
        filtered_df = df[(df["team"] == team_name) & (df['is_starter'] == True)]
        team_dict = {}

        for _, row in filtered_df.iterrows():
            category = self.categorize_position(row["position"])
            if category not in team_dict:
                team_dict[category] = []
            team_dict[category].append(row["player"])

        return team_dict

    def get_league_table_sync(self, team, start_date=None, end_date=None):
        async def fetch():
            async with aiohttp.ClientSession() as session:
                understat = Understat(session)
                return await understat.get_league_table(self.league, self.season, start_date=start_date, end_date=end_date, title=team)

        return asyncio.run(fetch())

    def get_team_players_sync(self, team, start_date, end_date):
        async def fetch():
            async with aiohttp.ClientSession() as session:
                understat = Understat(session)
                return await understat.get_team_players(team, year, end_date=date)

        return asyncio.run(fetch())



    def get_data(self, date: date, home, away, h_lineup=None, a_lineup=None, ref=None):

        table = self.get_league_table_sync(home, end_date=date)
        return table
```
